```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

def plot_chi_square(degrees_of_freedom):
    x = np.linspace(0, 20 * degrees_of_freedom, 1000)
    chi2_pdf = chi2.pdf(x, degrees_of_freedom)
    
    plt.plot(x, chi2_pdf, label=f'Chi-Square PDF (df={degrees_of_freedom})')
    plt.xlabel('Chi-Square Value')
    plt.ylabel('Probability Density')
    plt.title('Chi-Square Distribution')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_chi_square(4)  # Change the degrees of freedom as needed
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Given observed data
observed = np.array([[116, 107],
                     [0, 5],
                     [0.0, 4.7]])

# Calculate expected frequencies
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
total = observed.sum()

expected = np.outer(row_totals, col_totals) / total

# Perform chi-square test
chi2, p, dof, expected_table = chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p)

# Conclusion based on the result
alpha = 0.05  # Significance level

if p < alpha:
    print("Reject the null hypothesis.")
    print("There is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is no significant association between the variables.")
```
